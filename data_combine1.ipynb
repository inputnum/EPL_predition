```python
# 모든 팀들의 전 경기 데이터 추출후 저장

import aiohttp
import asyncio
import pandas as pd
from understat import Understat

async def main():
    async with aiohttp.ClientSession() as session:
        understat = Understat(session)
        teams = await understat.get_teams("EPL", 2017)
        all_results = []

        for team in teams:
            team_name = team['title']
            results = await understat.get_team_results(team_name, 2017)
            team_data = pd.DataFrame(results)
            team_data['home_team'] = team_data.apply(lambda row: row['h']['title'] if row['side'] == 'h' else row['a']['title'], axis=1)
            team_data['away_team'] = team_data.apply(lambda row: row['a']['title'] if row['side'] == 'h' else row['h']['title'], axis=1)
            team_data['home_xG'] = team_data.apply(lambda row: row['xG']['h'] if row['side'] == 'h' else row['xG']['a'], axis=1)
            team_data['away_xG'] = team_data.apply(lambda row: row['xG']['a'] if row['side'] == 'h' else row['xG']['h'], axis=1)
            team_data['datetime'] = pd.to_datetime(team_data['datetime'])
            team_data['result'] = team_data['result']
            team_data['datetime'] = team_data['datetime'].dt.strftime('%Y-%m-%d')
            all_results.append(team_data)

    # 각 팀의 결과를 병합
    df_all_teams = pd.concat(all_results, keys=[team['title'] for team in teams])

    # 필요한 컬럼만 선택하여 출력
    df_all_teams = df_all_teams.droplevel(0)[['home_team', 'away_team', 'home_xG', 'away_xG', 'datetime', 'result']]
    print(df_all_teams)



loop = asyncio.get_event_loop()
loop.run_until_complete(main())

df_all_teams.to_csv('~/Documents/ds_study/final_data/final_data2017.csv', index=False)
```

```python
file1_path = '~/Documents/ds_study/final_data/combined_file2.csv'
file2_path = '~/Documents/ds_study/final_data/combined_file3.csv'

df1 = pd.read_csv(file1_path)
df2 = pd.read_csv(file2_path)
```

```python
# 각 파일의 HomeTeam 및 AwayTeam 열의 고유한 값들 가져오기
unique_teams_df1 = set(df1['HomeTeam'].unique()) | set(df1['AwayTeam'].unique())
unique_teams_df2 = set(df2['HomeTeam'].unique()) | set(df2['AwayTeam'].unique())

# 두 파일의 HomeTeam 및 AwayTeam 열의 고유한 값들을 통합
all_unique_teams = unique_teams_df1 | unique_teams_df2

# 통합된 고유한 값들을 기준으로 각 파일의 HomeTeam 및 'AwayTeam' 열 최신화
df1['HomeTeam'] = df1['HomeTeam'].map(lambda x: x if x in all_unique_teams else None)
df1['AwayTeam'] = df1['AwayTeam'].map(lambda x: x if x in all_unique_teams else None)
df2['HomeTeam'] = df2['HomeTeam'].map(lambda x: x if x in all_unique_teams else None)
df2['AwayTeam'] = df2['AwayTeam'].map(lambda x: x if x in all_unique_teams else None)


df1
```
